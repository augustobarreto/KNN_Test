```python
# Primeiro, criaremos um dicionário para mapear os 22 atributos, a cada indíce. O primeiro atributo pode ser b,c,x,f,k,s a quem deve-se atribuir 0 para n, 1 para c, 2 para x, 3 para f, 4 para k e 5 para s. Fazemos o primeiro elemento de preparation ser preparation[1] = 'bcxfks' assim, podemos fazer a troca pela indexação da string. Ex: preparation[1].index('c') = 2, preparation[5].index('y') = 3
preparation = {1: 'bcxfks', 2: 'fgys', 3: 'nbcgrpuewy', 4: 'tf', 5: 'alcyfmnps', 6: 'adfn', 7: 'cwd', 8: 'bn',
               9: 'knbhgropuewy', 10: 'et', 11: 'bcuezr?', 12: 'fyks', 13: 'fyks', 14: 'nbcgopewy', 15: 'nbcgopewy',
               16: 'pu', 17: 'nowy', 18: 'not', 19: 'ceflnpsz', 20: 'knbhrouwy', 21: 'acnsvy', 22: 'glmpuwd'}
```

```python
import math
# Número de vizinhos
knn = int(input())
n = int(input())

#Ler o arquivo csv com os dados
import csv

with open('mushroom2.csv', 'r') as f:
    reader = csv.reader(f)
    data_as_list = list(reader)
```

```python
# Número de linhas da matriz x_train
n_train = n;
# Numero de linhas da matriz x_test
n_test = len(data_as_list) - n;

i = 0
j = 0
x_train = [[0 for j in range(22)] for i in range(n)];
y_train = [0 for j in range(n)] ;

```

```python
# Ler os n_train's input()'s, para formar a matriz de treinamento x_train, criada acima. Essa matriz terá n_train linhas e 22 colunas.
for i in range(n_train):
    j = 0;
    for j in range (22):
        x_train[i][j] = data_as_list[i][j];   
```

```python
# A função modeling recebe a matriz x_train e devolve a matriz df. A matriz df é o conjunto de dados númericos da matriz x_train e tem tamanho igual a matriz x_train.
def modeling(n_train, x_train):
    df = []
    # Esse laço percorrerá as k linhas da matriz x_train e criará uma lista temporária df_temp.
    for k in range(n_train):
        i = 1
        df_temp = []
        # Esse laço percorrerá cada coluna da linha fixada k. E cada valor numerico de preparation[i].index(x_train[k][i-1]) será adicionado a lista df_temp. Assim o k está fixo e o i variando de 1 até 22. Tem i-1 pois o primeiro valor de x_train é x_train[0][0] enquanto o dicionário começa por preparation[1]. 
        while i != 23:
            df_temp.append(preparation[i].index(x_train[k][i - 1]))
            i += 1
        # Por fim, com o fim do laço while, teremos df_temp sendo uma matriz com uma linha e 22 colunas dos valores númericos da linha fixada de x_train. Assim adicionamos df_temp a df. Df terá tamanho igual a x_train.
        df.append(df_temp)
    return df
```

```python
# Salvamos a matriz de treinamento com os valores númericos como sendo df_x
df_x = modeling(n_train, x_train)
```

```python
# df_x é uma matriz numerica, a função transposed() vai pegar a matriz df_x e devolver a matriz df_x_transposed, que é a matriz transposta de df_x.
def transposed(n_train, df_x):
    df_x_transposed = []
    # Criamos a lista df_x_transposed e rodamos um loop para percorrer de 1 até 22 e uma lista temporária, df_temp, que terá sempre n_train colunas e 1 linhas. 
    for k in range(22):
        j = 0
        df_temp = []
        # Aqui, percorremos cada linha de df_x, a matriz de treinamento,com a k coluna fixada, adicionado os valores a lista df_temp. 
        while j != n_train:
            df_temp.append(df_x[j][k])
            j += 1
        # Com o fim do laço while, adicionamos a lista df_temp à lista df_x_transposed, que no final do processo terá 22 linhas e n_train colunas.
        df_x_transposed.append(df_temp)
    return df_x_transposed
```

```python
# Salvamos a matriz transposta de df_x como transposed_X. 
transposed_x = transposed(n_train, df_x)
```

```python
# Na função mean(), vamos calcular a media de cada linha da matriz transposed_x, a media de cada atributo.
def mean(transposed_x):
    mean = []
    # Um laço que percorre cada linha de transposed_x e adiciona a lista mean o valor da media de cada linha i fixada.
    for i in transposed_x:
        mean.append(sum(i) / len(i))
    # mean vai ser uma matriz com uma linha e 22 colunas. 
    return mean
```

```python
# Salvamos a matriz das medias da matriz tranposta como mean_x. 
mean_x = mean(transposed_x)
```

```python
# Na função standard_deviation(), vamos calcular o desvio padrão de calinha da matriz transposed_x, o desvio pardão de cada atributo. 
def standard_deviation(mean, transposed):
    standard_deviation = []
    l = 0
    # Neste laço, faremos i variar nas 22 linhas de transposed.
    for i in transposed:
        variance = 0
        # Neste laço, pegamos cada elemento j, da coluna i, subtraímos a média da respectiva linha e depois elevamos ao quadrado e salvamos como variance.
        for j in i:
            variance = variance + ((j - mean[l]) ** 2)
        # Aqui, salvamos na lista standar_deviation a raiz quadrada de variance dividido pelo tamanho da lista i, para cada linha i fixada.
        standard_deviation.append(math.sqrt(variance / len(i)))
        l += 1
        # A matriz standar_deviation terá uma linha e 22 colunas.
    return standard_deviation
```

```python
# Salvamos como standar_deviation_x, a lista dos valores do desvio padrão de cada linha da matriz transposed_x. 
standard_deviation_x = standard_deviation(mean_x, transposed_x)
```

```python
# Na função normalize(), vamos normalizar a matriz de treinamento, df_x.
def normalize(df_x, standard_deviation_x, mean):
    # Neste laço, percorreremos cada linha da matriz df_x. Com um contador y, que vai de 0 até 22.
    for i in df_x:
        y = 0
        k = 0
        #Neste laço, vamos percorrer cada coluna da linha i fixada. 
        for k in range(22):
            # Se o desvio padrão da coluna y for 0, o elemento df_x[i][k] será igual a 0.
            if standard_deviation_x[y] == 0:
                i[k] = 0
            # Se o desvio padrão da coluna y não for 0, Então o elemento i[k], será igual a i[k] = df_x[i][k] = (df_x[i][k] - mean_x[y]) / standar_deviation_x[y]. Assim, cada elemento da linha i de df_x será subtraído pela média da coluna em que está e depois, dividido pelo desvio padrão da coluna em que está.
            else:
                i[k] = (i[k] - mean_x[y]) / standard_deviation_x[y]
            y += 1
    return df_x
```

```python
#Salvamos como df_normalized_x, a matriz df_x normalizada. 
df_normalized_x = normalize(df_x, standard_deviation_x, mean_x)
```

```python
#Aqui vamos ler os y_train's input()'s.
for i in range(n_train):
        y_train[i] = data_as_list[i][22];  
```

```python
#Aqui vamos criar um dicionário para associar cada linha da matriz treinamento a um valor "p" ou "e" da lista y_train. Vai ficar: {1 : 'p', 2: 'e', ... , n_train: 'p'}
dic_df_x = {}
i = 0
for x in df_x:
    dic_df_x[i] = y_train[i]
    i += 1
```

```python
#Aqui vamos ler os n_test valores da matriz teste.
i = 0
x_test = [[0 for j in range(22)] for i in range(n_test)];
y_test = [0 for j in range(n_test)] ;

for i in range(n_test):
    j = 0;
    y_test[i] = data_as_list[n_train+i][22];  
    for j in range (22):
        x_test[i][j] = data_as_list[n_train+i][j];   
        
```

```python
#Aqui vamos transformá-la numa matriz númerica.
df_x_test = modeling(n_test, x_test)

#Aqui vamos normalizar a matriz númerica de teste.
df_x_test_normalized = normalize(df_x_test, standard_deviation_x, mean_x)



```

```python
#Aqui vamos calcular a distancia euclidiana de cada linha i da matriz normalizada de teste, com cada uma das linhas da matriz normalizada de treinamento.
distances_total = []
#Vamos iterar cada uma das i linhas da matriz normalizada de teste. Com um contador, k, e uma lista temporário distances2.
for i in df_x_test_normalized:
    distances2 = []
    k = 0
    #Aqui, iterar cada k linha da matriz normalizada de treinamento. Criando uma segunda lista temporário, distances1 e um contador y.
    for k in df_normalized_x:
        distances1 = []
        y = 0
        # Agora, vamos fazer a distancia, termo a termo, entre a linha i fixada da matriz de teste normalizada, com a linha k fixada da matriz de treinamento normalizada.
        while y != len(k):
            #Adicionamos cada resultado da linha i pela linha k à matriz distances1, que terá uma linha e 22 colunas.
            distances1.append((i[y] - k[y]) ** 2)
            y += 1
        #Adicionamos a distances2, a raiz quadrada do somátorio da lista distance1. A matriz distances2 terá uma linha e 22 colunas.
        distances2.append(math.sqrt(sum(distances1)))
    #Por fim, adicionamos cada resultado de distances2, a matriz distances_total. A matriz distances_total terá n_test linhas e 22 colunas.
    distances_total.append(distances2)
```

```python
#Aqui vamos associar as distancias ao dicionário dic_df_x para sabermos se o vizinho é comestível, 'p', ou venenoso, 'e'.     
a = []
#Vamos percorrer o tamanho de n_test, pois n_test é igual a quantidade de linhas da matriz distances_total. E criamos uma lista temporária.
for i in range(n_test):
    lista = []
    #Aqui vamos descobrir cada rótulo do j vizinho mais próximo. knn é o primeiro input do trabalho.
    for j in range(knn):
        #Vamos adicionar a lista 'lista' o rótulo dos vizinhos.
        #distances_total[i].index(sorted(distances_total[i])[j]): 
        #a lista sorted(distances_total[i]) é a lista distance_total[i] ordenada, 
        #.index(sorted(distances_total[i])[j]) retorna o indice do j-esimo elemento da linha i da matriz distances_total. Lembrando q 0 <=0 j <= knn. Assim j=0 é o vizinho mais proximo, j = knn é o knn-esimo vizinho mais proximo.
        #dic_df_x[distances_total[i].index(sorted(distances_total[i])[j])] >> retorna o rótulo 'p' ou 'e' da linha de teste i dos j vizinhos mais próximos. 
        #Lista terá uma linha e knn colunas.
        lista.append(dic_df_x[distances_total[i].index(sorted(distances_total[i])[j])])
    #a terá n_test linhas e knn colunas.
    a.append(lista)
```

```python
b = [];

for i in range(n_test):
    
    #Aqui checamos se a maioria dos knn vizinhos mais proximos da linha i da matriz test são comestiveis, se sim imprimimos p, se não imprimimos e.
    if a[i].count('p') > (knn / 2):
        print('p')
        b.append('p');
    else:
        print('e')
        b.append('e');
```

```python
#Comparar Y Classificado com Y Verdadeiro
acum = 0;
i = 0;
for i in range(n_test):
    if b[i] == y_test[i]:
        acum = acum + 1;
    
```
